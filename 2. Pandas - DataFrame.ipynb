```python
# apply
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def dis(row):
    a = row['1st point']
    b = row['2nd point']
    return ((a[0]-b[0])**2 + (a[1]-b[1])**2)**0.5
```

```python
points_df.apply(dis,axis=1)
```

```python
points_df['distance']=points_df.apply(dis,axis=1)
```
